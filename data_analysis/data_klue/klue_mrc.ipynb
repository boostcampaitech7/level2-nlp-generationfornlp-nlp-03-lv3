```python
from datasets import load_dataset

# KLUE-MRC 데이터셋 불러오기
dataset = load_dataset("klue", "mrc")
```

```python
df_train = dataset['train']
```

```python
set(df_train['source'])
```

```python
pd_train = df_train.to_pandas()
```

```python
subjects = ['경제', '교육산업', '국제', '부동산', '사회', '생활', '책마을']
pd_subjects = pd_train[pd_train['news_category'].isin(subjects)]
pd_subjects
```

```python
pd_subjects.to_csv('kluemrc_subjects.csv')
```
