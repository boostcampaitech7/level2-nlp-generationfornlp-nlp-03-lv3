## test id칼럼 정렬

```python
import pandas as pd

df = pd.read_csv('../resources/raw/test_reformat.csv')
df
```

```python
# id의 숫자 부분을 기준으로 정렬
df = df.sort_values(by='id', key=lambda x: x.str.extract(r'(\d+)').squeeze().astype(int))
df
```

```python
df.to_csv('test_reformat_sorted.csv')
```

## source 칼럼에 source/subject

```python
df['id'] = df['id'].str.extract(r'(\d+)', expand=False).astype(int)
```

```python
# 범위에 따라 source 값 할당
def assign_source(row):
    if 0 <= row['id'] <= 424:
        return 'KCSAT'
    elif 425 <= row['id'] <= 513:
        return 'KMMLU/korean_history'
    elif 514 <= row['id'] <= 661:
        return 'MMMLU/european_history'
    elif 662 <= row['id'] <= 670:
        return 'MMMLU/geography'
    elif 671 <= row['id'] <= 680:
        return 'MMMLU/government_and_politics'
    elif 681 <= row['id'] <= 770:
        return 'MMMLU/macroeconomics'
    elif 771 <= row['id'] <= 816:
        return 'MMMLU/microeconomics'
    elif 817 <= row['id'] <= 974:
        return 'MMMLU/psychology'
    elif 975 <= row['id'] <= 1151:
        return 'MMMLU/us_history'
    elif 1152 <= row['id'] <= 1379:
        return 'MMMLU/word_history'
    elif 1380 <= row['id'] <= 2899:
        return 'KLUE-MRC'
    else:
        return None  # 범위에 해당하지 않는 경우

# apply 함수로 source 컬럼 생성
df['source'] = df.apply(assign_source, axis=1)

df
```

```python
df.to_csv('test_reformat_with_source_and_subject.csv')
```
