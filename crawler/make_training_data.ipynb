```python
import re
def clean(paragraph):
    paragraph = paragraph.replace('&rarr;', '->')
    paragraph = re.sub(r'&[^;]+;', '', paragraph)
    paragraph = paragraph.replace('<strong>', '')
    paragraph = paragraph.replace('</strong>', '')
    paragraph = paragraph.replace('<strong/>', '')
    paragraph = paragraph.replace('<br>', '')
    paragraph = paragraph.replace('</br>', '')
    paragraph = paragraph.replace('<br/>', '')
    paragraph = paragraph.replace('</p>', '')
    paragraph = paragraph.replace('<p style="text-align:right">', '')
    paragraph = paragraph.replace('<p class="txtrt">', '')
    paragraph = paragraph.replace('<p class="txt_rt">', '')
    paragraph = paragraph.replace('<p style="border:1px solid #ccc;padding:10px;">', '')
    paragraph = paragraph.replace('<p style="text-align:right;">', '')
    paragraph = paragraph.replace('<p style="text-align:center;">', '')
    paragraph = paragraph.replace('<span class="p_start"></span>', '')
    paragraph = paragraph.replace('<div class="ex_box">', '')
    paragraph = paragraph.replace('<p class="pl10">', '')
    paragraph = paragraph.replace('  ', '')
    paragraph = paragraph.replace('   ', '')
    return paragraph
```
