```python
import requests

# DeepL API 키와 URL
api_key = 'c36fe977-0279-4fba-9b3a-275babc2feb9:fx'
url = 'https://api-free.deepl.com/v2/translate'

# 번역 함수
def translate_text(text, target_lang='KO'):
    params = {
        'auth_key': api_key,
        'text': text,
        'target_lang': target_lang
    }
    response = requests.post(url, data=params)
    result = response.json()

    # 무료 요금제 한도 초과 시 오류 처리
    if 'message' in result and 'Quota exceeded' in result['message']:
        print("번역 한도를 초과했습니다. 더 이상 번역할 수 없습니다.")
        return None  # 번역을 멈추고 None 반환

    return result['translations'][0]['text']
```

```python
# 데이터셋 번역
df_world_history['question'] = df_world_history['prompt'].apply(lambda x: translate_text(x, 'KO'))
df_world_history['choice_A'] = df_world_history['A'].apply(lambda x: translate_text(x, 'KO'))
df_world_history['choice_B'] = df_world_history['B'].apply(lambda x: translate_text(x, 'KO'))
df_world_history['choice_C'] = df_world_history['C'].apply(lambda x: translate_text(x, 'KO'))
df_world_history['choice_D'] = df_world_history['D'].apply(lambda x: translate_text(x, 'KO'))
df_world_history['choice_E'] = df_world_history['E'].apply(lambda x: translate_text(x, 'KO'))

# 번역된 데이터를 새 CSV로 저장
df_world_history.to_csv('sat_world_history.csv', index=False)
```

## world_history 후처리

### reformat 형태로 만들기

```python
df_world_history = pd.read_csv('sat_world_history.csv')
```

```python
# 'choices' 칼럼 생성
df_world_history['choices'] = df_world_history[['choice_A', 'choice_B', 'choice_C', 'choice_D', 'choice_E']].apply(lambda row: row.tolist(), axis=1)

# 'answer' 칼럼을 A, B, C, D, E에서 1, 2, 3, 4, 5로 변환
answer_mapping = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5,
                  'A.': 1, 'B.': 2, 'C.': 3, 'D.': 4, 'E.': 5, '.': 0}
df_world_history['answer'] = df_world_history['answer'].map(answer_mapping).astype(int)

# 불필요한 칼럼 삭제
df_world_history = df_world_history[['id', 'subject', 'question', 'choices', 'answer']]

# 결과를 새 CSV로 저장
df_world_history.to_csv('sat_world_history_processed.csv', index=False)
```

### answer에 text 답변 집어넣기

```python
import ast


df_world_history = pd.read_csv('sat_world_history_processed.csv')

# answer 번호를 텍스트로 변환하는 함수
def map_answer(row):
    choices = ast.literal_eval(row['choices'])  # 문자열을 리스트로 변환
    answer_index = row['answer'] - 1  # answer은 1부터 시작, 리스트 인덱스는 0부터 시작
    return choices[answer_index]

# answer를 텍스트로 변환
df_world_history['answer'] = df_world_history.apply(map_answer, axis=1)
```

## world_history 증강 데이터셋 (with paragraph)

### paragraph max token EDA

```python
df = pd.read_csv('../../resources/processed/train_reformat_with_source_and_subject.csv')
```

```python
df_sat = df[df['source'].isin(['MMMLU/word_history', 'MMMLU/us_history', 'MMMLU/european_history'])]
```

```python
# 문자열 길이 계산
df_sat['paragraph_length'] = df_sat['paragraph'].apply(lambda x: len(str(x)))

# max, mean, min 계산
max_length = df_sat['paragraph_length'].max()
mean_length = df_sat['paragraph_length'].mean()
min_length = df_sat['paragraph_length'].min()

# 결과 출력
print(f"Max length: {max_length}")
print(f"Mean length: {mean_length:.2f}")
print(f"Min length: {min_length}")
```

```python
from transformers import AutoTokenizer

# Hugging Face 토크나이저 로드
tokenizer = AutoTokenizer.from_pretrained("../../saved_model")

# 토큰 수 계산
df_sat['token_count'] = df_sat['paragraph'].apply(lambda x: len(tokenizer.tokenize(str(x))))

# max, mean, min 계산
max_tokens = df_sat['token_count'].max()
mean_tokens = df_sat['token_count'].mean()
min_tokens = df_sat['token_count'].min()

# 결과 출력
print(f"Max token count: {max_tokens}")
print(f"Mean token count: {mean_tokens:.2f}")
print(f"Min token count: {min_tokens}")
```

```python
df_world = pd.read_csv('sat_world_history_paragraph.csv')

# 토큰 수 계산
df_world['token_count'] = df_world['paragraph'].apply(lambda x: len(tokenizer.tokenize(str(x))))

# max, mean, min 계산
max_tokens = df_world['token_count'].max()
mean_tokens = df_world['token_count'].mean()
min_tokens = df_world['token_count'].min()

# 결과 출력
print(f"Max token count: {max_tokens}")
print(f"Mean token count: {mean_tokens:.2f}")
print(f"Min token count: {min_tokens}")
```
