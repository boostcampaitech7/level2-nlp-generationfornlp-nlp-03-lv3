## source 칼럼에 source/subject

```python
import pandas as pd

df = pd.read_csv('../resources/raw/train_reformat.csv')
```

```python
df['id'] = df['id'].str.extract(r'(\d+)', expand=False).astype(int)
df
```

```python
# 범위에 따라 source 값 할당
def assign_source(row):
    if 425 <= row['id'] <= 513:
        return 'KMMLU/korean_history'
    elif 514 <= row['id'] <= 661:
        return 'MMMLU/european_history'
    elif 662 <= row['id'] <= 670:
        return 'MMMLU/geography'
    elif 672 <= row['id'] <= 680:
        return 'MMMLU/government_and_politics'
    elif 682 <= row['id'] <= 770:
        return 'MMMLU/macroeconomics'
    elif 771 <= row['id'] <= 816:
        return 'MMMLU/microeconomics'
    elif 817 <= row['id'] <= 974:
        return 'MMMLU/psychology'
    elif 975 <= row['id'] <= 1150:
        return 'MMMLU/us_history'
    elif 1152 <= row['id'] <= 1379:
        return 'MMMLU/word_history'
    elif 1380 <= row['id'] <= 2899:
        return 'KLUE-MRC'
    else:
        return None  # 범위에 해당하지 않는 경우

# apply 함수로 source 컬럼 생성
df['source'] = df.apply(assign_source, axis=1)

df
```

```python
# 각 source별 개수와 비율 구하기
target_counts = df['source'].value_counts()
target_ratios = df['source'].value_counts(normalize=True)

# 결과 출력
result_df = pd.DataFrame({
    'Count': target_counts,
    'Ratio': target_ratios
})

print(result_df)
```

```python
df.to_csv('train_reformat_with_source_and_subject.csv')
```
